Se extrae del dataset de `train` una matriz más pequeña con 2000 filas de muestra aleatoria, para observar facil y rapidamente los datos:

```python
resum = train.sample(n=2000) 
resum
```

--------------------------------
Se comparan algunas de las columnas numéricas unas con otras para tratar de reconocer patrones.

```python
cols_num = ['cont0', 'cont1', 'cont2', 'cont8', 'cont9', 'cont10']
sns.set()
sns.pairplot(resum[cols_num])
```

Se alcanza a ver un poco lo sintético de los datos, pero se reconoce un patrón de correspondencia entre, por ejemplo, las columnas `cont2-cont1`, `cont1-cont8` o `cont2-cont8`. Puede ser que los datos de estas 3 columnas tratasen de las mismas cuestiones en los datos originales.

--------------

Ahora observemos las columnas categóricas:

```python
from collections import Counter
print(Counter(train['cat0']))
print(Counter(train['cat1']))
print(Counter(train['cat2']))
print(Counter(train['cat3']))
print(Counter(train['cat4']))
print(Counter(train['cat5']))
print(Counter(train['cat6']))
print(Counter(train['cat7']))
```

```python
from collections import Counter
col_cat = ['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7', 'cat8',
        'cat9', 'cat10', 'cat11', 'cat12', 'cat13', 'cat14', 'cat15', 'cat16',
        'cat17', 'cat18']

result = Counter()
for x in col_cat:
    result = result + Counter(train[x])

result.most_common(1)
```

Si bien hay un mar de datos absolutamente regados y aleatorios a la vista en esta muestra, se observa que la variable que más se repite es la correspondiente a la letra `A`, por tanto es posible que sea la de mayor peso para la decisión de asignacion al `target` de `0` o `1`.
